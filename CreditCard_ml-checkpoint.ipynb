```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X1=data.drop(['Class'],axis=1)
y1=data['Class']
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=123)
lr=LogisticRegression()
model_lr= lr.fit(X1_train, y1_train)
prediction1 = model_lr.predict(X1_test)
accuracy_score(y1_test, prediction1)
```

```python
X2 = data.drop(['Class'],axis=1)
y2= data['Class']
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2 ,test_size=0.3, random_state=123)
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
model_rfc = rfc.fit(X2_train, y2_train)
prediction2 = model_rfc.predict(X2_test)
from sklearn.metrics import accuracy_score
accuracy_score(y2_test, prediction2)
```

```python
from sklearn.tree import DecisionTreeRegressor
X3 = data.drop(['Class'],axis=1)
y3 = data['Class']
dt = DecisionTreeRegressor()
X3_train, X3_test,y3_train, y3_test = train_test_split(X3, y3, test_size = 0.3,random_state=123)
model_dt = dt.fit(X3_train, y3_train)
prediction3 = model_dt.predict(X3_test)
accuracy_score(y3_test, prediction3)
```

```python
report = classification_report(y1_test, prediction1)
print(report)
```

```python
report = classification_report(y2_test, prediction2)
print(report)
```
